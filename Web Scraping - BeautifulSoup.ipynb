```python
def getOnePage(page_link):
    ##access webpage
    page_respond = requests.get(page_link)
    # initialize beautifulsoup
    page_content = BeautifulSoup(page_respond.content, "html.parser")
    
    container = page_content.find(name= "div", attrs = {"id": "feature-product-grid"})
    
    ##get product name
    products_name = container.find_all(name ="div", attrs = {"class": "product-title"})
    list_name = []
    for name in products_name:
        list_name.append(name.text)
        
    # get price
    final_price = container.find_all(name ="p", attrs = {"class": "final-price"})
    list_finalprice = []
    for price in final_price:
        list_finalprice.append(int(price.text.replace(".","").replace("đ", "")))
        
    # get description
    prod_des = container.find_all(name ="div", attrs = {"class": "popover-top-content"})
    list_des = []
    for des in prod_des:
        list_des.append(des.text)
        
    #get photos
    prod_img = container.find_all(name ="img")
    list_linkimg = []
    for image in prod_img:
        if image.has_attr("data-src"):
            list_linkimg.append(image["data-src"])
            
    # get photo name
    list_img = []
    for link in list_linkimg:
        img_name = link.split("/")[-1]
        list_img.append(img_name)
    
        with open("img/" + img_name, "wb") as f:
            image_response = requests.get(link)  ## lay du lieu tu link
            f.write(image_response.content)
    # save as dataframe
#         df_temp = pd.DataFrame({
#         "name": list_name,
#         "price": list_finalprice,
#         "Description": list_des,
#         "image": list_img
#     })  LOIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIIII
    
    next_page = container.find(name = "a", attrs = {"class": "btn_next ty-pagination__item ty-pagination__btn ty-pagination__next cm-history "})
    next_link = ""
    if next_page:
        next_link = next_page["href"]
    return df_temp, next_link
```

```python
## call function
page_link = "https://www.nguyenkim.com/tu-lanh/"
list_df = []
while True:
    df_temp, next_link = getOnePage(page_link)
    list_df.append(df_temp)
    if next_link == "":
        break
    link = next_link
df    
```
