## Price

```python
product_price = driver.find_elements(by = By.CSS_SELECTOR, value = "span.price-num")
```

```python
list_price = []
for price in product_price:
    p = price.text
    list_price.append(p)
    
```

```python
valueToBeRemoved = 'đ'
try:
    while True:
        list_price.remove(valueToBeRemoved)
except ValueError:
    pass
```

```python
list_price = [x.strip() for x in list_price if x.strip()]
```

```python
list_price
```

```python
len(list_price)
```

```python
deposit_price = product_price = driver.find_elements(by = By.CSS_SELECTOR, value = "p.nt-prepaid-installment>strong")
```

```python
list_dprice = []
for price in deposit_price:
    gia = price.text.replace('đ','')
    list_dprice.append(gia)

list_dprice = [x.strip() for x in list_dprice if x.strip()]
```

## Photo

```python
product_images = driver.find_elements(by = By.CSS_SELECTOR, value = "img.ty-pict.cm-image")
```

```python
len(product_images)
```

```python
list_imglink = [p.get_attribute("src") for p in product_images]
```

```python
valueToBeRemoved1 = 'https://www.duchuymobile.com/images/empty.jpg'
try:
    while True:
        list_imglink.remove(valueToBeRemoved1)
except ValueError:
    pass
```

```python
list_imglink
```

```python
list_image = []
for image_link in list_imglink:
    image_name = "img-selenium/" + image_link.split("/")[-1]
    image = requests.get(image_link, allow_redirects = True)
    with open(image_name, "wb") as f:
        f.write(image.content)
    list_image.append(image_link.split("/")[-1])
```
